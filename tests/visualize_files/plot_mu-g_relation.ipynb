```python
import torch
import matplotlib.pyplot as plt
import numpy as np

x = torch.linspace(0, 1, 1000)

y_default = (-0.0155 * np.log(x) + 0.0244).clamp(min=0.049, max=0.1)
y_default_rev = (-0.0155 * np.log(x) + 0.0244).clamp(min=0.01)
y_linear =  (-0.05 * x + 0.1)
y_const_010 = 0.1 + x*0
y_const_005 = 0.05 + x*0
y_const_001 = 0.01 + x*0
y_gst225 = -0.1841* (x**3) + 0.4204* (x**2) - 0.3134 * x + 0.08465
y_msr = 0.0513*np.exp(-4.9751*x**0.5803) + 0.0069

# plt.plot(x,y_default, label='default', linewidth=2, linestyle='--', color='k')
plt.plot(x,y_default_rev, label = 'logarithm', linewidth=2, color='k')
plt.plot(x,y_linear, label = 'linear', linewidth=2, color='dodgerblue')
# plt.plot(x,y_const_010, label = '0.1', linewidth=2, color='darkorange')
# plt.plot(x,y_const_005, label = '0.05', linewidth=2, color='g')
# plt.plot(x,y_const_001, label = '0.01', linewidth=2, color='m')
# plt.plot(x,y_gst225, label = 'GST225', linewidth=3, color='r')
plt.plot(x,y_msr, label = 'MSR', linewidth=3, color='r')

plt.legend(fontsize=12, loc='upper right', bbox_to_anchor=(1.0, 1.0))
# plt.legend(fontsize=12, loc='lower left', bbox_to_anchor=(0.0, 0.0))
plt.xlabel('Conductance (norm.)', fontsize=18, fontweight='medium')
plt.ylabel('Mean of drift coefficient (mu)', fontsize=18)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
# plt.xscale('log')
# plt.xlim(0,0.05)
# plt.ylim(0,0.15)
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import torch

x = torch.linspace(0, 1, 1000)
a_values = [1.0, 2.0, 5.0]
for a in a_values:
    y =  (-0.0155 * np.log((x*a)+0.00762) + 0.0244).clamp(min=0.0244)
    plt.plot(x, y, label=f'a={a}', linewidth=2)

plt.legend()
# plt.xscale('log')
plt.xlim(0,1)
plt.ylim(0,0.1)
plt.grid(True)
plt.show()
```
