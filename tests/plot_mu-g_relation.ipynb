```python
import torch
import matplotlib.pyplot as plt
import numpy as np

x = torch.linspace(0, 1, 1000)

y_default = (-0.0155 * np.log(x) + 0.0244).clamp(min=0.049, max=0.1)
y_default_rev = (-0.0155 * np.log(x) + 0.0244).clamp(min=0.01)
y_linear =  (-0.05 * x + 0.1)
y_const_010 = 0.1 + x*0
y_const_005 = 0.05 + x*0
y_const_001 = 0.01 + x*0

plt.plot(x,y_default, label='default', linewidth=2, linestyle='--', color='k')
plt.plot(x,y_default_rev, label = 'default_rev(log)', linewidth=1, color='k')
plt.plot(x,y_linear, label = 'linear', linewidth=2, color='dodgerblue')
plt.plot(x,y_const_010, label = '0.1', linewidth=2, color='darkorange')
plt.plot(x,y_const_005, label = '0.05', linewidth=2, color='g')
plt.plot(x,y_const_001, label = '0.01', linewidth=2, color='m')

plt.legend(fontsize=10)
plt.xlabel('Conductance (norm.)', fontsize=14)
plt.ylabel('Mean of drift coefficient (mu)', fontsize=14)
plt.grid(True)
plt.show()
```
