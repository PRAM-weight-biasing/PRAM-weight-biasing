```python
import os
import numpy as np
import pandas as pd

file_path = './260421_test_acc.xlsx'

df_raw = pd.read_excel(file_path, sheet_name='Sheet1', engine='openpyxl')
df_raw = df_raw.loc[:, ~df_raw.columns.str.contains("Unnamed:")]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))

# drift_noise_scale별로 그룹핑하여 플롯
for noise_scale in df_raw['drift_noise_scale'].unique():
    data = df_raw[df_raw['drift_noise_scale'] == noise_scale]
    plt.plot(data['inf_time'], data['accuracy'], marker='o', label=f'drift_noise_scale: {noise_scale}')

plt.xscale('log')
plt.xlabel('Inference Time (log scale)', fontsize=12)
plt.ylabel('Accuracy', fontsize=12)
plt.title('Accuracy vs Inference Time', fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
